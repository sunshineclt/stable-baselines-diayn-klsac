```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = ""
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import csv
%matplotlib inline
```

```python
basedir = "log/diayn/HalfCheetah-v2_5/"
files = [basedir + "openai_1000000/0.monitor.csv", 
         basedir + "openai_1200000/0.monitor.csv", 
         basedir + "openai_1400000/0.monitor.csv", 
         basedir + "openai_1600000/0.monitor.csv", ]
```

```python
all_rewards = np.zeros([4, 20])
for file_idx, file in enumerate(files):
    with open(file, "r") as f:
        next(f)
        reader = csv.DictReader(f)
        for row_idx, row in enumerate(reader):
            all_rewards[file_idx, row_idx] = float(row["r"])
```

```python
plt.figure(figsize=[12, 8])
for i in range(15, 20):
    plt.plot(all_rewards[:, i], label=i)
plt.legend()
```
